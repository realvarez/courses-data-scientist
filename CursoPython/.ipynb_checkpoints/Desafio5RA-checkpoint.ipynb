## Ejercicio 1

```python
import pandas as pd
import numpy as np
```

```python
INPUT_BASE_PATH = "Data/"
INCIDENTS_FILE_NAME = 'incidents.pkl'
OFFICERS_FILE_NAME = 'officers.pkl'
SUBJECTS_FILE_NAME = 'subjects.pkl'
```

### a - Carga los datos y crea un DataFrame con cada uno de ellos

```python
incidents = pd.read_pickle(INPUT_BASE_PATH+INCIDENTS_FILE_NAME)
officers = pd.read_pickle(INPUT_BASE_PATH+OFFICERS_FILE_NAME)
subject = pd.read_pickle(INPUT_BASE_PATH+SUBJECTS_FILE_NAME)
display("INCIDENTS", incidents.head())
display("OFFICERS", officers.head())
display("SUBJECT", subject.head())
```

### b - Genera una tabla que contenga la unión de las 3 tablas

```python
# Mergeamos 
df = incidents.merge(officers, how = "left", on = "case_number", suffixes=(None,"_officer"))\
            .merge(subject, how = "left", on = "case_number", suffixes=('_officer',"_subject"))\
            
```

### c - Verifica si hay filas duplicadas; si es así, elimínalas

```python
df[df.duplicated(keep=False)]
```

```python
#No se tienen filas duplicadas
df = df.drop_duplicates()
```

### d - ¿Cuántos sujetos de género F hay en el DataFrame resultante?

```python
print(f"Se tienen {df.gender_subject.value_counts()['F']} mujeres sospechosas")
```

### e - ¿En cuántos números de caso hay por lo menos una sospechosa que sea mujer?

```python
case_number_female_suspect = df[df.gender_subject == "F"].case_number
print(f"Se tienen {case_number_female_suspect.drop_duplicates().count()} casos diferentes donde hay por lo menos un sospechoso mujer")
```

### f - Tabla Pivote
Genera una tabla pivote que muestre en las filas el género del oficial y en las columnas el 
género del subject. ¿Cómo interpretas los valores que muestra esta vista?

```python
df_genders = df[['gender_officer', 'gender_subject']].copy()
df_genders['num'] = 1
df_genders.pivot_table(values='gender_subject', index=['gender_officer'], columns='gender_subject', aggfunc='sum')
```

## Ejercicio 2

### a -Carga los datos y crea un DataFrame con ellos

```python
INPUT_BASE_PATH = "Data/"
DS_JOBS_FILENAME = 'Cleaned_DS_Jobs.csv'
```

```python
import pandas as pd
df = pd.read_csv(INPUT_BASE_PATH+DS_JOBS_FILENAME)
df.head()
```

### b - Utiliza la siguiente lista de valores que serán considerados como nulos 
    
    ["na", "NA", -1, "0", "-1", "null", "n/a", "N/A", "NULL"]
    (hint: utiliza el método replace para reemplazar los valores indicados por np.nan)

```python
import numpy as np
lista_nulos = ["na", "NA", -1, "0", "-1", "null", "n/a", "N/A", "NULL"]
df = df.replace(to_replace = lista_nulos, value=np.nan)
df.head()
```

### c - Elimina todas las filas con datos faltantes. (hint: utiliza el método .dropna()

```python
df = df.dropna()
df.head()
```

### d - A partir de la columna “Salary Estimate”, genera dos columnas: 
    Salario Estimado Minimo
    Salario Estimado Máximo. 
    (hint: Utiliza el método apply sobre la colu### mna.)

```python
df['Salary Estimate'][0].split('-', 2)
```

```python
def separe_min_salary_estimate(salary_estimate: str):
    return int(salary_estimate.split('-', 2)[0])
    
def separe_max_salary_estimate(salary_estimate: str):
    return int(salary_estimate.split('-', 2)[1])

df['Salary Estimate Min'] = df['Salary Estimate'].apply(separe_min_salary_estimate)
df['Salary Estimate Max'] = df['Salary Estimate'].apply(separe_max_salary_estimate)
```

### e - Realiza la recodificación de la columna Size con los valores de la siguiente tabla: (hint: utilice reemplazo con diccionario usando el método replace sobre la columna.)

|Size                    |Descripción              |
|------------------------|-------------------------|
|10000+ employees        |Mega Empresas            |
|5001 to 10000 employees |Grandes Empresas         |
|1001 to 5000 employees  |Medianas Empresas        |
|201 to 500 employees    |Pequeñas Empresas        |
|51 to 200 employees     |Pequeñas Grandes Empresas|
|501 to 1000 employees   |Microempresas            |
|Unknown                 |Empresas sin Información |


```python
dict_size_company = {
    '10000+ employees': 'Mega Empresas',
    '5001 to 10000 employees': 'Grandes Empresas',
    '1001 to 5000 employees': 'Medianas Empresas',
    '201 to 500 employees': 'Pequeñas Empresas',
    '51 to 200 employees': 'Pequeñas Grandes Empresas',
    '501 to 1000 employees': 'Microempresas',
    'Unknown': 'Empresas sin Información'
}
```

```python
df.Size = df.Size.apply(dict_size_company.get)
```

###  f - Genera una tabla pivote 
Debe mostrar la media del salario estimado mínimo y 
la media del salario estimado máximo por tamaño de empresa.
 (hint: utiliz 
pd.pivot_table para generar la vista adecuada con las columnas generadas.)

```python
df.pivot_table(values=['Salary Estimate Min', 'Salary Estimate Max'], columns="Size", aggfunc='mean')
```

```python
print(df[df.Size == 'Grandes Empresas']['Salary Estimate Min'].mean())
print(df[df.Size == 'Mega Empresas']['Salary Estimate Min'].mean())
```
