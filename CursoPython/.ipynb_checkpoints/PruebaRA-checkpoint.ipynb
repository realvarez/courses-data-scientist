### Investigacion de relacion entre tablas

```python
print("order number unico?", len(set(df_orders.orderNumber)) == df_orders.shape[0])
len(set(df_orderdetails.orderNumber)) == df_orderdetails.shape[0] 
# Tiene orderNumbers repetidos, son los productos pedidos en cada orden 
```

```python
print("customer number unico?", len(set(df_orders.customerNumber)) == df_orders.shape[0])
len(set(df_customers.customerNumber)) == df_customers.shape[0]
# Catalogo de clientes, se relaciona 1:m con la tabla de ordenes
```

```python
print("product code unico?", len(set(df_orderdetails .productCode)) == df_orderdetails.shape[0])
len(set(df_products.productCode)) == df_products.shape[0]
# Catalogo de productos, se relaciona 1:m con la tabla de detalle de ordenes
```

```python
print("EmployeeNumber unico en customers?",len(set(df_customers.salesRepEmployeeNumber)) == df_customers.shape[0])
len(set(df_employees.employeeNumber )) == df_employees.shape[0]
# Catalogo de empleados, se relaciona 1:m con la tabla de detalle de customers
```

## Cruce de tablas

```python
df_merged = df_orders.merge(
    df_orderdetails, how="left", on="orderNumber", validate="1:m"
).merge(
    df_customers, how="left", on="customerNumber", validate="m:1"
).merge(
    df_products, how="left", on="productCode", validate="m:1"
).rename(
    columns={"salesRepEmployeeNumber": "employeeNumber"}
).merge(
    df_employees, how="left", on="employeeNumber", validate="m:1"
)

df_merged.head()
```

```python
print(df_merged.columns)
```

## Agrega columnas

```python
df_merged['venta'] = df_merged.quantityOrdered * df_merged.priceEach
df_merged['costo'] = df_merged.quantityOrdered * df_merged.buyPrice
df_merged['ganancia'] = df_merged.venta - df_merged.costo
df_merged.head(10)
```

## ¿Cuál fue el total de ventas por línea de productos?

```python
df_merged[['productLine', 'venta']].pivot_table(index='productLine', values='venta', aggfunc='sum', margins=True)
```

## ¿Cuántos clientes distintos hicieron compras?


```python
# Devido a que estoy cruzando los numeros de las ordenes con los clientes, todos los clientes que tengo tienen por lo menos una orden
num_clientes_distintos = len(set(df_merged[['orderNumber','customerNumber']].customerNumber))
print(f"El numero de clientes distintos que hicieron una compra es: {num_clientes_distintos}")
```

## ¿Existen clientes que aún no han hecho ninguna compra? ¿Cuántos son?

```python
df_customers_and_orders = df_customers.merge(df_orders, how="left", on="customerNumber", validate="1:m")[['customerNumber', 'orderNumber']]

clientes_sin_compra = set(df_customers_and_orders[pd.isna(df_customers_and_orders['orderNumber'])]['customerNumber'])

print(f"El numero de clientes que no han hecho compras es: {len(clientes_sin_compra)}")
```

## Generacion de reportes

### ¿Cuáles fueron los 10 clientes que reportan mayores ventas brutas en dinero durante el año 2005?

```python
df_filtrado_2005 = filter_by_date(df_merged, 'orderDate', '01-01-2005', '31-12-2005')
df_ventas_customer_2005 = generate_report(
    df=df_filtrado_2005, 
    filas=['customerNumber', 'customerName'], 
    columnas=None, 
    valores='venta', 
    agg_funct='sum'
)
cant_insertados = save_in_database(
    df_ventas_customer_2005.sort_values(by='venta', ascending=False).head(10), 
    table='top_10_clientes_2005', 
    engine=engine, 
    if_exist='replace'
)
print(f'Se insertaron {cant_insertados} registros en la base de datos')
```

### ¿Cuál fue el top 10 de artículos más vendidos durante el año 2005? (considerando cantidad neta)

```python
df_filtrado_2005 = filter_by_date(df_merged, 'orderDate', '2005-01-01', '2005-12-31')
df_reporte = generate_report(df_filtrado_2005, ['productCode', 'productName', 'venta', 'costo', 'ganancia'], None, 'quantityOrdered', agg_funct='sum')
cant_insertados = save_in_database(
    df_reporte.sort_values(by='quantityOrdered', ascending=False).head(10),
    'top_10_productos_2005',
    engine,
    'replace'
)
print(f'Se insertaron {cant_insertados} registros en la base de datos')
```
