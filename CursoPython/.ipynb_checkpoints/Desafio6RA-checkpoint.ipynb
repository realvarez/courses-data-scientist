```python
import pandas as pd
import numpy as np
```

```python
INPUT_BASE_PATH = "Data/desafio6"
SALES_DATA_FILE = 'US_Regional_Sales_Data.xlsx'
INPUT_SALES_FILE_PÄTH = f"{INPUT_BASE_PATH}/{SALES_DATA_FILE}"
```

```python
sales_file_sheet_names =  pd.ExcelFile(INPUT_SALES_FILE_PÄTH).sheet_names
sales_file_sheet_names
```

```python
dataframes = []
for sheet in sales_file_sheet_names:
    dataframes.append(pd.read_excel(INPUT_SALES_FILE_PÄTH, sheet_name=sheet))
```

```python
display(dataframes[0].head())
display(dataframes[1].head())
display(dataframes[2].head())
display(dataframes[3].head())
display(dataframes[4].head())
```

```python
df_base = pd.DataFrame()
for el in dataframes:
    if df_base.empty:
        df_base = el
    else:
        df_base = df_base.merge(el, how="left")
df_base.head()
```

```python
df_base['ProcurementDays'] = df_base.OrderDate-df_base.ProcuredDate
df_base['ShippingDays'] = df_base.ShipDate-df_base.OrderDate
df_base['DeliveryDays'] = df_base.DeliveryDate-df_base.ShipDate
df_base['CustomerDays'] = df_base.ShippingDays+df_base.DeliveryDays
df_base.head()
```

```python
etiquetas  =['0 to 15 days',
             '15 to 30 days',
             '30 to 45 days',
             '45 to 60 days',
             '60 to 75 days',
             '75 to 90 days']

intervalos = [0,15,30,45,60,75,90]

df_base['CustomerDaysInterval'] = pd.cut(df_base.CustomerDays.dt.days, intervalos, labels = etiquetas)
```

```python
df_base[['Sales Team', 'CustomerDaysInterval']].head()
```

```python
df_base[''] = 1
df_reporte_1 = df_base[
['Sales Team', 'CustomerDaysInterval', '']
].pivot_table(
    index='Sales Team', columns="CustomerDaysInterval", aggfunc="count", margins=True, observed=True
)
df_reporte_1
```

```python
df_base['GrossMargin'] = df_base['Order Quantity'] * (df_base['Unit Price'] * (1 - df_base['Discount Applied']) - df_base['Unit Cost'])
df_base['GrossMargin']
```

```python
percent  =[0, 5, 10, 15, 20]
intervalos = [-1000000, 0,100, 1000, 10000,100000]

df_base['CommissionsPercentage'] = pd.cut(df_base.GrossMargin, intervalos, labels = percent)
df_base['CommissionsAmount'] = (df_base.GrossMargin * df_base.CommissionsPercentage.astype('int')) / 100
df_base['NetMargin'] = df_base.GrossMargin - df_base.CommissionsAmount
df_base['NetMargin'] = df_base['NetMargin'].apply(lambda x: x if x > 0 else 0)
df_base
```

```python
df_reporte_2 = df_base[['Sales Team', 'CommissionsAmount', 'GrossMargin', 'NetMargin']].pivot_table(
    index='Sales Team', values=['CommissionsAmount', 'GrossMargin', 'NetMargin'], aggfunc="sum", margins=True
).round(2)
df_reporte_2
```
