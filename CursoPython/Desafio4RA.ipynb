```python
import pandas as pd
import math
import numpy as np
```

```python
INPUT_FILE_DIR = "resumen_resultados_astronautas.csv"
OUTPUT_FILE_DIR = "astronautas_calificados.csv"
IMC_MAX = 24.9
IMC_MIN = 18.5
MIN_PROMEDIO = 87
```

```python
df = pd.read_csv(INPUT_FILE_DIR)
print(df.shape)
df.head()
```

```python
df['altura'] = df.altura.map(lambda x: x/100)
df['IMC'] = pd.Series(
    row.peso / math.pow(row.altura ,2)
    for row in df.itertuples()
)
df['promedio_evaluaciones'] = pd.Series(
    np.mean([row.evaluacion_fisica, row.evaluacion_psicologica, row.evaluacion_adaptativa, row.evaluacion_medica])
    for row in df.itertuples()
)
print(df.shape)
df.head()
```

```python
df = df[np.logical_and(df.IMC.between(IMC_MIN, IMC_MAX), df.promedio_evaluaciones >= MIN_PROMEDIO)]
print(df.shape)
df.head()
```

```python
df.califica = 'Si'
df.reset_index(inplace=True, drop=True)
df.head()
```

```python
df.to_csv(OUTPUT_FILE_DIR)
```
